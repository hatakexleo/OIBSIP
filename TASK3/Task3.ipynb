```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('carprice.csv')
df
```

```python
df.info()
```

```python
df.drop(['car_ID','CarName'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
for i in df:
    #print(df[f'{i}'].dtype)
    if df[f'{i}'].dtype==object:
        #print('i')
        df[f'{i}']=enc.fit_transform(df[[f'{i}']])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


for i in df:
    plt.plot(df[f'{i}'],df.price,'*')
    plt.ylabel("price")
    plt.xlabel(str(i))
    plt.show()
```

```python
for i in df:
    plt.boxplot(df[f'{i}'])
    plt.title(str(i))
    plt.show()
```

```python
l=['fueltype','aspiration','carbody','enginelocation','wheelbase','carlength','carwidth','enginetype','cylindernumber','enginesize','stroke','compressionratio','horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']
df.columns
```

```python
def remove_outlier(s):
    sorted(s)
    q1=s.quantile(0.25)
    q3=s.quantile(0.75)
    iqr=q3-q1
    low=q1-(1.5*iqr)
    high=q3+(1.5*iqr)
    return low,high
```

```python
import numpy as np
for i in l:
    low,high=remove_outlier(df[f'{i}'])
    df[f'{i}']=np.where(df[f'{i}']>high,high,df[f'{i}'])
    df[f'{i}']=np.where(df[f'{i}']<low,low,df[f'{i}'])
```

```python
for i in df:
    plt.boxplot(df[f'{i}'])
    plt.title(str(i))
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
for i in df:
    ss.fit(df[[f'{i}']])
    df[[f'{i}']]=ss.transform(df[[f'{i}']])
```
